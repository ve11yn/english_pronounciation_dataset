```python
import os
import ffmpeg
```

```python
base_input = '../1. m4a_audio'
base_output = '../2. wav_audio + text'

for speaker in os.listdir(base_input):
    speaker_path = os.path.join(base_input, speaker)

    if os.path.isdir(speaker_path):
        output_folder = os.path.join(base_output, speaker)
        os.makedirs(output_folder, exist_ok=True)

        for file in os.listdir(speaker_path):
            if file.endswith(".m4a"):
                input_path = os.path.join(speaker_path, file)
                output_path = os.path.join(
                    output_folder,
                    file.replace(".m4a", ".wav")
                )

                (
                    ffmpeg
                    .input(input_path)
                    .output(output_path, ac=1, ar=16000)
                    .run(overwrite_output=True)
                )
```
